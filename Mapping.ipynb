```python
d = dict()
```

```python
f = open("products.csv", "r")
for s in f:
    for i in range(len(s)):
        if s[i] == ',':
            p_id = int(s[:i])
            p = s[i+1:len(s)-1]
            d[p_id] = p
            break
```

```python
bigStr = ""
file = "tr-75k.csv"
f = open(file, "r")
for s in f:
    l = s.split(',')
    l = l[1:]
    for n in l:
        n = int(n.replace(" ", "").replace("\n", ""))
        bigStr += d[n] + ","
    bigStr = bigStr[0:len(bigStr) -1] + "\n"
f = open(file, "w")
f.write(bigStr)
```
